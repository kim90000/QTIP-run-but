```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

model_id = "relaxml/Llama-3.1-8b-Instruct-QTIP-2Bit"
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(model_id, device_map="auto")

prompt = "What is AI?"

# **1. استخدام معاملات التوليد (Temperature و Top-k)**
temperature = 0.7  # يمكنك تجربة قيم أخرى مثل 0.5, 0.9, 1.2
top_k = 30        # يمكنك تجربة قيم أخرى مثل 10, 50, None (لإيقاف top-k)

# **2. استخدام chat_template إذا كان النموذج نموذج تعليمات/محادثة**
if tokenizer.chat_template is not None:
    messages = [{"role": "user", "content": prompt}]
    inputs = tokenizer.apply_chat_template(messages, return_tensors="pt").to(model.device) # تأكد من نقل المدخلات إلى جهاز النموذج
else:
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device) # تأكد من نقل المدخلات إلى جهاز النموذج

# **توليد النص مع معاملات التوليد**
outputs = model.generate(
    **inputs,
    max_length=50,
    temperature=temperature,
    top_k=top_k,
    do_sample=True # تفعيل أخذ العينات (ضروري لاستخدام temperature و top_k)
)

print(tokenizer.decode(outputs[0], skip_special_tokens=True))

# **3. (اختياري) تفعيل TF32 إذا كنت تستخدم FP32 وأجهزة NVIDIA تدعم TF32**
torch.set_float32_matmul_precision('high') # يتم تفعيله بشكل افتراضي في بداية البرنامج إذا كان ذلك مناسبًا

# **4. (اختياري) تجربة torch.compile (لأداء أفضل في حالات معينة)**
# إذا كنت تقوم بتوليد متكرر أو تسلسلات أطول، يمكنك تجربة تجميع دالة التوليد.
# في هذا المثال البسيط، قد لا يكون له تأثير كبير.
```
